```python
# selecting random 'society name'
society_name = df['society'].value_counts().index[0]
society_name
```

```python
# applying regular expression to fetch the only society name

# importing regular expression
import re

# regular expression
pattern = "\d+(\.+\d)?\s+?★"

# fetching the required details
clean_society_name = re.sub(pattern, "", society_name)
print(f"Clean society name: \n{clean_society_name}")

# removing any while spaces avaiable
clean_society_name = clean_society_name.strip()
print(f"\nRemoved any availabel white spaces:\n{clean_society_name}")

# changing the name to lower case
print(f"\nChanging the society name to lower case:\n{clean_society_name.lower()}")
```

```python
# applying the regular expression on complete column to get the society name
df['society'] = df['society'].apply(lambda name: re.sub("\d+(\.+\d)?\s+?★", "", str(name)).strip()).str.lower()
```

```python
# unique price values in the dataset
df.price.value_counts()
```

```python
# filtering out non-numerical entries in price column
filtered_prices = df.loc[~df.price.isna() & df.price.str.contains('^\D'), 'price']

# fetching the unique values
non_numerical_values = set(filtered_prices.tolist())


# checking the rows in the dataframe
price_mask = df.price.isin(non_numerical_values)

# printing the count
print(f"\nTotal rows in the dataset, which contains non numerical entries in price column: {df[price_mask].shape[0]}")

# dataframe view
df[price_mask]
```

```python
# considering only rows in the dataset, which contains the price in 'price column'
df = df[~df.price.isin(non_numerical_values)]

# shape of the dataset
print(f"Shape of the dataset: {df.shape}")
```

```python
# few prices in the dataset are in crore and few are in lac
# keeping the price the same for each row

def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
```

```python
# applying the above function on the price column
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
# checking the random five rows after updating the 'price' column
df.sample(5)
```

```python
# plotting a graph to show the price bins in the dataset

# we will the divide the price in 5 bins
bin_count = 5

# considering the unique price values from the dataset
price_values = df.price.value_counts().index.sort_values(ascending=True)

# splitting the prices in buckets
price_bins = pd.cut(price_values, bin_count, include_lowest=True)

# plotting the graph for price buckets
plt.figure(figsize=(8, 6))
price_bins.value_counts().sort_values(ascending=True).plot(kind='barh', color='#FF5833')
plt.xlabel('Frequency Number')
plt.ylabel('Price Buckets')
plt.title('Flat prices in price buckets')
plt.show()
```

```python
# price_per_sqft
# keeping the only price values and removing the extra text from 'price_per_sqft' column
df['price_per_sqft'] = df.price_per_sqft.str.split("/").str.get(0).str.replace("₹", '').str.replace(',', '').str.strip().astype('float')
```
