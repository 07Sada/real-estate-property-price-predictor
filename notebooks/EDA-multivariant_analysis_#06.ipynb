```python
# sector analysis
import re
# Group by 'sector' and calculate the average price
avg_price_per_sector = df.groupby('sector')['price'].mean().reset_index()

# Function to extract sector numbers
def extract_sector_number(sector_name):
    match = re.search(r'\d+', sector_name)
    if match:
        return int(match.group())
    else:
        return float('inf')  # Return a large number for non-numbered sectors

avg_price_per_sector['sector_number'] = avg_price_per_sector['sector'].apply(extract_sector_number)

# Sort by sector number
avg_price_per_sector_sorted_by_sector = avg_price_per_sector.sort_values(by='sector_number')

# Create the heatmap
fig, ax = plt.subplots(figsize=(5, 25))  # Adjust figure size for better readability
sns.heatmap(
    avg_price_per_sector_sorted_by_sector.set_index('sector')[['price']],
    cmap="Set2",  # Use the specified color palette
    annot=True,  # Show data values inside each heatmap cell
    fmt=".2f",  # Format as floats with two decimal places
    linewidths=0.5,  # Adjust grid line width between cells
    cbar_kws={"label": "Average Price"}  # Add label to the colorbar
)

# Customize the plot
ax.set_title("Average Price per Sector (Sorted by Sector Number)", fontsize=14)
ax.set_xlabel("Average Price", fontsize=12)  # Adjust label font sizes
ax.set_ylabel("Sector", fontsize=12)
plt.xticks(rotation=0)  # No rotation for x-axis labels (assuming sector names are short)
plt.tight_layout()  # Adjust spacing to prevent overlapping elements

plt.show()
```
